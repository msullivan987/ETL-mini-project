```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store filepath in a variable
csv_path= "netflix_titles.csv"
netflix_df = pd.read_csv(csv_path)
netflix_df.head()
```

```python
new_netflix_df = netflix_df[["show_id","title","release_year","rating"]].copy()
new_netflix_df
```

```python
new_netflix_df["title"] = new_netflix_df["title"].apply(lambda x: x.lower())
new_netflix_df
```

```python
#Connecting to local database
rds_connection_string = "postgres:postgres@localhost:5432/bechdel_netflix_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_netflix_df.to_sql(name='netflix_data', con=engine, if_exists='append', index=False)
```
